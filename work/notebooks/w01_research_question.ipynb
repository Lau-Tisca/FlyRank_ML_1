## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

* **What decision does this improve?** It optimizes the prioritization decision: *"Which specific page in our inventory should an editor review and update first to capture the highest return on effort?"*
* **Who acts on the output, and what do they do?** A content editor or SEO specialist acts on the output. They inspect the top-ranked pages alongside human-readable reason codes (e.g., `stale_visible_page`, `thin_visible_page`) and manually refresh, expand, prune, or monitor the page.
* **What does a wrong answer cost?** * *False Positive (Type I Error):* The system prioritizes an healthy page, resulting in wasted editor hours auditing content that didn't require intervention.
    * *False Negative (Type II Error):* The system misses a high-exposure page experiencing severe decay, leading to irreversible loss of search equity and traffic.
* **Why does data or ML help at all?** Fixed hand-tuned rule thresholds fail to map the non-linear, tangled relationship between impression volume, seasonal blips, age tiers, and engagement rates. A learned ranking model maps these multi-system signals more dynamically.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Load the starter anonymized dataset directly from the raw GitHub URL
url = 'https://raw.githubusercontent.com/Lau-Tisca/FlyRank_ML/main/data/raw/content_refresh_anonymized.csv'
df = pd.read_csv(url)

# Calculate baseline metrics
total_inventory = len(df)
decline_rate = (df['trend_direction'] == 'down').mean() * 100

# Identify high-exposure stale pages (impressions >= 500 and age >= 180 days)
high_exposure_stale = df[(df['impressions_90d'] >= 500) & (df['content_age_days'] >= 180)]
stale_count = len(high_exposure_stale)

print(f"--- Starter Dataset Insights ---")
print(f"1. Total Page Inventory: {total_inventory:,} items")
print(f"2. Raw Proxy Decline Rate (trend_direction == 'down'): {decline_rate:.2f}%")
print(f"3. High-Exposure Stale Candidates (Imps >= 500 & Age >= 180d): {stale_count:,} pages")
```
